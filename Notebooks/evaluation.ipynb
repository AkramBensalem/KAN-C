```python

# Assuming 'results' contains the KAN forecast and 'model1_dataset' holds the test data.
# Assuming 'model1_dataset['test_input']' holds the input features for the test set.

# Access the true values (targets) from the test set
true_values = model1_dataset['test_label'].cpu().numpy()  # Convert to numpy for plotting

# Step 1: Make predictions on the test set using the KAN model
with torch.no_grad():
    test_predictions = kan_model(model1_dataset['test_input'])

# Step 2: Convert tensors to NumPy arrays for plotting
kan_forecasts = test_predictions.cpu().numpy()

# Step 3: Use complex_expression function to create symbolic_forecast
symbolic_forecast = []
symbolic_forecast1 = []
with torch.no_grad():
    for i in range(model1_dataset['test_input'].shape[0]):
        # Extract each input set from the test dataset and pass it to complex_expression
        x = model1_dataset['test_input'][i]
        symbolic_result = complex_expression(x)  # Pass each sample to the complex expression
        symbolic_forecast.append(symbolic_result.cpu().numpy())  # Append to list after converting to numpy
        symbolic_result1 = complex_expression1(x)  # Pass each sample to the complex expression
        symbolic_forecast1.append(symbolic_result1.cpu().numpy())  # Append to list after converting to numpy

# Convert symbolic_forecast to a NumPy array for easier handling and plotting
symbolic_forecast = np.array(symbolic_forecast)
symbolic_forecast1 = np.array(symbolic_forecast1)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

# Assuming `kan_forecasts`, `symbolic_forecast`, and `true_values` are already defined and are NumPy arrays.

# Calculate MAE for KAN forecast
mae_kan = mean_absolute_error(true_values, kan_forecasts)

# Calculate MAE for Symbolic forecast
mae_symbolic = mean_absolute_error(true_values, symbolic_forecast)
# Calculate MAE for Symbolic forecast
mae_symbolic1 = mean_absolute_error(true_values, symbolic_forecast1)

# Calculate RMSE for KAN forecast
rmse_kan = np.sqrt(mean_squared_error(true_values, kan_forecasts))
# Calculate RMSE for symbolic forecast
rmse_symbolic = np.sqrt(mean_squared_error(true_values, symbolic_forecast))
# Calculate RMSE for symbolic forecast
rmse_symbolic1 = np.sqrt(mean_squared_error(true_values, symbolic_forecast1))
# Display the results
print(f"Mean Absolute Error (MAE) for KAN forecast: {mae_kan}")
print(f"Mean Absolute Error (MAE) for Symbolic forecast: {mae_symbolic}")
print(f"Mean Absolute Error (MAE) for Symbolic forecast1: {mae_symbolic1}")
print(f"Root Mean Square Error (RMSE) for KAN forecast: {rmse_kan}")
print(f"Root Mean Square Error (RMSE) for Symbolic forecast: {rmse_symbolic}")
print(f"Root Mean Square Error (RMSE) for Symbolic forecast: {rmse_symbolic1}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata

# Step 1: Extract features and calculate errors
id_values = true_values[:, 0]  # i_d values (from true_values)
iq_values = true_values[:, 1]  # i_q values (from true_values)

kan_errors = np.abs(kan_forecasts - true_values)  # Absolute errors for KAN
symbolic_errors = np.abs(symbolic_forecast - true_values)  # Absolute errors for symbolic

# Step 2: Calculate MAE and RMSE for KAN and Symbolic forecasts
kan_mae = np.mean(kan_errors, axis=1)  # Mean absolute error for KAN
symbolic_mae = np.mean(symbolic_errors, axis=1)  # Mean absolute error for Symbolic

# Calculate RMSE
kan_rmse = np.sqrt(np.mean(kan_errors**2, axis=1))  # Root mean square error for KAN
symbolic_rmse = np.sqrt(np.mean(symbolic_errors**2, axis=1))  # Root mean square error for Symbolic

# Step 3: Create a grid for i_d and i_q
id_range = np.linspace(id_values.min(), id_values.max(), 100)
iq_range = np.linspace(iq_values.min(), iq_values.max(), 100)
ID, IQ = np.meshgrid(id_range, iq_range)

# Step 4: Interpolate MAE and RMSE values onto the grid
kan_mae_grid = griddata((id_values, iq_values), kan_mae, (ID, IQ), method='linear')
symbolic_mae_grid = griddata((id_values, iq_values), symbolic_mae, (ID, IQ), method='linear')

kan_rmse_grid = griddata((id_values, iq_values), kan_rmse, (ID, IQ), method='linear')
symbolic_rmse_grid = griddata((id_values, iq_values), symbolic_rmse, (ID, IQ), method='linear')

# Step 5: Plot contour plots for MAE and RMSE
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# KAN MAE contour plot
contour1 = axes[0, 0].contourf(ID, IQ, kan_mae_grid, cmap='coolwarm', levels=50)
axes[0, 0].set_title('KAN Model MAE')
axes[0, 0].set_xlabel('$i_d$')
axes[0, 0].set_ylabel('$i_q$')
fig.colorbar(contour1, ax=axes[0, 0])

# Symbolic MAE contour plot
contour2 = axes[0, 1].contourf(ID, IQ, symbolic_mae_grid, cmap='coolwarm', levels=50)
axes[0, 1].set_title('Symbolic Model MAE')
axes[0, 1].set_xlabel('$i_d$')
axes[0, 1].set_ylabel('$i_q$')
fig.colorbar(contour2, ax=axes[0, 1])

# KAN RMSE contour plot
contour3 = axes[1, 0].contourf(ID, IQ, kan_rmse_grid, cmap='coolwarm', levels=50)
axes[1, 0].set_title('KAN Model RMSE')
axes[1, 0].set_xlabel('$i_d$')
axes[1, 0].set_ylabel('$i_q$')
fig.colorbar(contour3, ax=axes[1, 0])

# Symbolic RMSE contour plot
contour4 = axes[1, 1].contourf(ID, IQ, symbolic_rmse_grid, cmap='coolwarm', levels=50)
axes[1, 1].set_title('Symbolic Model RMSE')
axes[1, 1].set_xlabel('$i_d$')
axes[1, 1].set_ylabel('$i_q$')
fig.colorbar(contour4, ax=axes[1, 1])

plt.tight_layout()
plt.show()

# Save the plot to a specific location
save_path = '../Images/paper_img/reality_vs_Kan_vs_symbolic_contour_grid.png'  # Change this path as needed
plt.savefig(save_path, format='png', dpi=300, bbox_inches='tight')
print("image saved at " + save_path)
```
