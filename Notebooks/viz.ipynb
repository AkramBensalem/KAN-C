```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
data = pd.read_csv('..\Data\Dataset_Electric_Motor.csv')

# Display the first few rows of the dataframe
print(data.head(50))
```

```python
data.describe()
```

```python

interval = 50
# Step 2: Create a plot for id_k1 and iq_k1
plt.figure(figsize=(12, 6))

# Plot id_k1 evolution
plt.plot(data.iloc[:interval].index, data['id_k1'].iloc[:interval], label=r'$i_{d,k+1}$', color='blue', marker='o')
# Plot iq_k1 evolution
plt.plot(data.iloc[:interval].index, data['iq_k1'].iloc[:interval], label=r'$i_{q,k+1}$', color='red', marker='x')
# Plot id_k1 evolution
plt.plot(data.iloc[:interval].index, data['id_k'].iloc[:interval], label=r'$i_{d,k}$', color='green', marker='o')
# Plot iq_k1 evolution
plt.plot(data.iloc[:interval].index, data['iq_k'].iloc[:interval], label=r'$i_{q,k}$', color='yellow', marker='x')

# Step 3: Add labels, title, and legend
#plt.xlabel('Index (Individuals or Time)')
# Hide x-axis values
plt.xticks([])  # This removes only the x-axis tick labels
plt.ylabel('Values')
plt.title(r'Zoomed view of $i_{d},i_{q}$ at instants k,k+1')
plt.legend()

# Step 4: Show the plot
plt.grid(True)
plt.show()
```
